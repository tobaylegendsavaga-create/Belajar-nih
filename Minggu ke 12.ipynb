```python
class Manusia :
    def __init__(self, nama):
        self.nama = nama

class Mahasiswa (Manusia):
    def __init__(self, nama, nim):
        super().__init_(nama):
        self.nim = nim

    def nama_nim (self):
        return self.nama, self.nim
    
manusia2 = Manusia ("agung")
mahasiswa2 = Mahasiswa ("lala", "123456")
print (mahasiswa2.nama.nim())
```
